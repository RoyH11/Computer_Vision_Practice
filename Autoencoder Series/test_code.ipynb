```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(10)
y = np.random.rand(10)
categories = [0, 1, 0, 1, 2, 2, 1, 0, 2, 1]  # Discrete values

plt.scatter(x, y, c=categories, cmap='tab20')
plt.colorbar()  # Show color legend
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(50)
y = np.random.rand(50)
values = np.linspace(0, 1, 50)  # Continuous values

plt.scatter(x, y, c=values, cmap='viridis')
plt.colorbar()  # Show the color gradient
plt.show()
```
